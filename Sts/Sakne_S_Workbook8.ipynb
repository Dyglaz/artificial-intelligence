–ø–æ–ø—ã—Ç–∞–µ—Ç—Å—è –∏–¥–µ–Ω—Ç–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –ø–æ—Ö–æ–∂–∏–µ —Ü–∏—Ñ—Ä—ã


```python
# –ù–∞—á–Ω–µ–º —Å –∏–º–ø–æ—Ä—Ç–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø–∞–∫–µ—Ç–æ–≤
import matplotlib.pyplot as plt
import seaborn as sns; sns. set()
import numpy as np
from sklearn.cluster import KMeans

# –∑–∞–≥—Ä—É–∑–∏–º –Ω–∞–±–æ—Ä —Ü–∏—Ñ—Ä –∏–∑ sklearn
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
# –≤—ã–ø–æ–ª–Ω–∏–º –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é
kmeans = KMeans(n_clusters = 10, random_state = 0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

–ü–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from pandas import DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import scipy.cluster.hierarchy as shc
sns.set(rc={'figure.figsize':(15,9)})
```

```python
X = np.array([[5,3], 
              [10,15],
              [15,12], 
              [24,10], 
              [30,45], 
              [85,70], 
              [71,80], 
              [60,78], 
              [55,52], 
              [80,91],])
df = pd.DataFrame(X)
df = df.rename(columns={0:'X', 1:'Y'})
for i in range(2, 10):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    clusters = kmeans.fit_predict(X)
    df[i]=clusters
sns.scatterplot(data=df, x='X',y='Y', hue=3, s=1000, alpha=1);
```

# –ó–∞–¥–∞–Ω–∏–µ:

–í—ã–ø–æ–ª–Ω–∏—Ç–µ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –¥–ª—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –∏—Ä–∏—Å–æ–≤ –§–∏—à–µ—Ä–∞. –í—ã–ø–æ–ª–Ω–∏—Ç–µ 
–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏. –ü–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.


```python
iris = load_iris()
X = DataFrame(iris.data)
X = X.rename(columns={0:'Sepal length',
                  1:'Sepal width',
                  2:'Petal length',
                  3:'Petal width'})
y = DataFrame({"target":iris.target})
for i in range(2, 10):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    clusters = kmeans.fit_predict(X)
    y[i]=clusters
df = pd.concat([X, y], axis=1)
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='Sepal length', y ='Sepal width', hue=3, s=250, alpha=1);
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Petal length', y ='Petal width', hue=3, s=250, alpha=1);
```

# –ü—Ä–∏–º–µ—Ä 1.1.2

```python
import matplotlib.pyplot as plt
import numpy as np
X = np.array(
    [[7,8], [12,20],[17,19], [26,15],[32,37],[87,75],[73,85], [62,80],[73,60], [87,96],])
labels = range(1, 11)
plt.figure(figsize = (10, 7))
plt.subplots_adjust(bottom = 0.1)
plt.scatter(X[:,0],X[:,1], label = 'True Position')
for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,xy = (x, y), xytext = (-3, 3),textcoords = 'offset points', ha = 'right', va = 'bottom')
plt.show()
```

–î–∞–ª–µ–µ –ø–æ—Å—Ç—Ä–æ–∏–º –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è —Ç–æ—á–µ–∫ –¥–∞–Ω–Ω—ã—Ö —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ 
Scipy

```python
from scipy.cluster. hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
linked = linkage(X, 'single')
labelList = range(1, 11)
plt.figure(figsize = (10, 7))
dendrogram(linked, orientation = 'top', labels = labelList,
    distance_sort ='descending',show_leaf_counts = True)
plt.show()
```

–î–∞–ª–µ–µ –Ω–∞–º –Ω—É–∂–Ω–æ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–ª–∞—Å—Å –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –∏ –≤—ã–∑–≤–∞—Ç—å –µ–≥–æ 
–º–µ—Ç–æ–¥ fit_predict –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –∫–ª–∞—Å—Ç–µ—Ä–∞.


```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
cluster.fit_predict(X)
plt.scatter(X[:,0],X[:,1], c = cluster.labels_, cmap = 'rainbow')
```
