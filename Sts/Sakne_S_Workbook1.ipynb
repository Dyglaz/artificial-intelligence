# 4.3.2 Задание

Дана функция. Создать массив из 10 значений
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный

```python
def func(x):
    a = m.sqrt(1 + m.exp(m.sqrt(x)) + m.cos(x) ** 2)
    b = abs(1 - m.sin(x) ** 3)
    return a/b

y = []
x = []
for i in range(1, 11):
    y.append(func(i))
    x.append(i)
plt.grid()
plt.plot(x, y, c = "r")
plt.fill_between(x, y)
plt.show()

plt.grid()
plt.scatter(x[:5:1], y[:5:1])
plt.show()
```

# 4.3.5 Задание

Создайте простейший калькулятор, включающий основные действия
для двух переменных

```python
print("Ноль в качестве знака операции"
      "\nзавершит работу программы"
      "\nпосчитать e^(x+y) - 1"
      "\nпосчитать sin(x+y) - 2"
      "\nпосчитать cos(x+y) - 3"
      "\nпосчитать x^y - 4")
while True:
    s = input("Знак (+,-,*,/,1,2,3,4): ")
    if s == '0':
        break
    if s in ('+', '-', '*', '/','1', '2', '3', '4'):
        x = float(input("x="))
        y = float(input("y="))
        if s == '+':
            print("%.2f" % (x+y))
        elif s == '-':
            print("%.2f" % (x-y))
        elif s == '*':
            print("%.2f" % (x*y))
        elif s == '/':
            if y != 0:
                print("%.2f" % (x/y))
            else:
                print("Деление на ноль!")
        elif s == '1':
            print("%.2f" % (m.pow(m.e,(x+y))))
        elif s == '2':
            print("%.2f" % (m.sin(x+y)))
        elif s == '3':
            print("%.2f" % (m.cos(x+y)))
        elif s == '4':
            print("%.2f" % (m.pow(x,y)))
    else:
        print("Неверный знак операции!")
```
