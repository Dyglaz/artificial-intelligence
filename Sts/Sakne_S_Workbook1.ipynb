# 4.3.2 –ó–∞–¥–∞–Ω–∏–µ

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
def func(x):
    a = m.sqrt(1 + m.exp(m.sqrt(x)) + m.cos(x) ** 2)
    b = abs(1 - m.sin(x) ** 3)
    return a/b

y = []
x = []
for i in range(1, 11):
    y.append(func(i))
    x.append(i)
plt.grid()
plt.plot(x, y, c = "r")
plt.fill_between(x, y)
plt.show()

plt.grid()
plt.scatter(x[:5:1], y[:5:1])
plt.show()
```
